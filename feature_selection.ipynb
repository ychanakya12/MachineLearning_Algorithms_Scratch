```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\ChanakyaY\Downloads\train.csv')
df.shape
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[: , -1]
```

```python
X.shape
```

```python
y.shape
```

```python
def normalise(X):
    u = X.mean(axis = 0)
    std = X.std(axis=0)
    X = (X-u)/ std
    return X
    
```

```python
X = normalise(X)
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features = SelectKBest()
```

```python
fit = best_features.fit(X,y)
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns , dfscores ],axis = 1)
featureScores.columns = ['Features','scores' ]
```

```python
featureScores
```

```python
featureScores = featureScores.sort_values(by = "scores",ascending = False)
featureScores
```

```python
top_10 = list(featureScores[:10]["Features"].values)
top_10
```

```python
# Model compartision
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
model.fit(X,y)
```

```python
scores = cross_val_score(model,X,y,cv=10)
scores.mean()
```

```python
scores = cross_val_score(model,X[top_10],y,cv=10)
scores.mean()
```
