```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons, make_circles, make_classification
```

```python
X,y = make_moons(n_samples = 10000 , noise = 0.15,random_state= 1 )
X1,y1 = make_circles(n_samples = 10000 , noise = 0.10,random_state= 1,factor = 0.5 )
X2,y2 = make_classification(n_samples = 10000 ,n_features = 2, n_redundant = 0 ,random_state= 1 )
```

```python
print(X.shape,y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c= y) # non linear data
plt.show()
```

```python
plt.scatter(X1[:,0],X1[:,1],c= y1) # non linear data
plt.show()
```

```python
plt.scatter(X2[:,0],X2[:,1],c= y2) # non linear data
plt.show()
```

## Model Building 

```python
from keras import Sequential # as we progree in sequential in neural network from the input data
from keras.layers import Dense # from every neuron to other there is a dense connection
```

```python
model = Sequential()
model.add(Dense(10,activation = 'relu',input_shape = (2,)))# input shape is given to first hidden layer only
model.add(Dense(5, activation = 'relu'))
model.add(Dense(1 , activation = 'sigmoid'))# output layer should be sigmoid bcz the final prob should in bw 0 and 1
model.compile(loss = 'binary_crossentropy',metrics = ['accuracy'])
model.summary()
```

```python
model.fit(X,y,batch_size = 64, epochs = 50, validation_split = 0.1)
```
